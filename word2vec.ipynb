```python
import pandas as pd
messages=pd.read_csv('/content/SMSSpamCollection',sep='\t',names=['label','message'])
messages
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

```

```python
corpus=[]
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)
```

```python
words = []

for text in corpus:
    sentences = sent_tokenize(text)
    for sentence in sentences:
        words.append(simple_preprocess(sentence))

```

```python
model = Word2Vec(
    sentences=words,
    vector_size=100,
    window=5,
    min_count=2,   # removes noise
    workers=4
)
```

```python
import numpy as np

def avg_word2vec(sentence):
    vectors = []
    for word in sentence:
        if word in model.wv.key_to_index:   # ✅ safest check
            vectors.append(model.wv[word])
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)
```
